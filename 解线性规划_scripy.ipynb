## 解线性规划

```python
import numpy as np
from scipy import optimize as op
```

```python
# 定义目标函数系数
c = np.array([2, 3, 1])


# 定义约束条件系数
A_ub = np.array([[-1, -4, -2],
                 [-3, -2, 0]])

B_ub = np.array([-8, -6])


# 定义决策变量范围
x1 = (0, None)
x2 = (0, None)
x3 = (0, None)

# 求解
res = op.linprog(c, A_ub, B_ub, bounds=(x1, x2, x3))
res
```
