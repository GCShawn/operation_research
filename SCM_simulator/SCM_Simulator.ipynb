前提假设
1. 1个月28天，4周，1周7天
2. 每个月的需求服从正态分布，SIGMA为10，需求以平均每周的方式到达。

使用方法：
1. 把某个月每周的调度计划写在SCM_Method.xlsx中（Dec 1001 = 59的意思是12月每周制作59个1001零件）
2. SCM_Simulator.xlsx包含所有题目参数，每月需求是从product_sales表中获得
3. simulate_the_demand.xlsx为仿真出的需求（自动生成）
4. simulate_the_product_after_make.xlsx为根据调度计划仿真的制造输出（自动生成）
5. simulate_result.xlsx为最后结果，包含累计仓储价值，最终库存，最终累计满足的需求，最终累计未满足需求，服务水平。

```python
import numpy as np
import pandas as pd
```

```python
DAYS_OF_MONTH = 28
DAYS_OF_WEEK = 7
WEEK_IN_MONTH = 4

SIGMA = 10                 #SIGMA
MIN_SIGMA_OF_ORINGIN = 8   #作为限制参数，以后或许会使用

FILE_PATH = r'SCM_Simulator.xlsx'
FILE_METHOD_PATH = r'SCM_Method.xlsx'

OUTPUT_FILE_PATH = r'simulate_the_demand.xlsx'
OUTPUT_FILE_PATH2 = r'simulate_the_product_after_make.xlsx'
OUTPUT_FILE_PATH3 = r'simulate_result.xlsx'

SHEET_NAME = r'product_sales'

dic_of_Feedstock = {'1mm Sheet':450,'3mm Plate':350,'6mm Plate':250,'100mm Bar':150}
dic_of_Product = {'1mm Sheet':800,'3mm Plate':700,'6mm Plate':600,'100mm Bar':450}
```

## 读取到的调度方法展示

```python
df_of_method = pd.read_excel(FILE_METHOD_PATH, SHEET_NAME, index_col = [0,1], header = 0)
dic = {"Dec":"1","Jan":"2","Feb":"3","Mar":"4","Apr":"5","May":"6","Jun":"7","Jul":"8","Aug":"9","Sep":"10"}

for i in df_of_method.columns:
    loc = df_of_method.columns.get_loc(i)
    if i in dic:
        for j in range(1,5):
            df_of_method.insert(loc,dic[i]+"_"+str(5-j),df_of_method[i])
        df_of_method.drop(columns=[i],inplace=True)
df_of_method
```

## 仿真方法给的80%成功率

```python
# 仿真80%成功率
max_x,max_y = df_of_method.shape
for i in range(max_x):
    for j in range(max_y-2):
        simulator = 0
        irret = df_of_method.iloc[i,j]
        for n in range(irret):
            a = np.random.rand()
            if a <= 0.8:#仿真成功率
                simulator += 1
        df_of_method.iloc[i,j] = simulator
df_of_method.to_excel(OUTPUT_FILE_PATH2,sheet_name=SHEET_NAME)#写入文件
df_of_method
```

## 按照仿真的需求和仿真的方法算出库存价值

```python
# df_of_product_sales 仿真出的每周需求
# df_of_method 仿真出的每周供货
#填充后2周的时间，确保unfulfill全部计算完成
df_of_method['10_1'] = 0
df_of_method['10_2'] = 0
df_of_product_sales['10_1'] = 0
df_of_product_sales['10_2'] = 0


df_of_calculating = pd.DataFrame().reindex_like(df_of_product_sales).fillna(0)
df_of_calculating.drop(df_of_calculating.columns, axis=1, inplace=True)

df_of_calculating['value_accumu'] = 0  # 计算所有物料的仓储成本
df_of_calculating['stock'] = 0  # 库存数量，由于order-make，所以没有考虑以零件的方式进行仓储情况

df_of_calculating['fullfill_demand'] = 0  # 计算满足了多少需求，未满足的需求就抛弃
df_of_calculating['unfulfill_demand'] = 0  # 记录未被满足的需求

df_of_calculating['demand_today'] = 0  # 每周一来到的需求
df_of_calculating['demand_wait_1W'] = 0  # 等待了1周的未被满足的需求
df_of_calculating['demand_wait_2W'] = 0  # 等待了2周的未被满足的需求，如果继续未被满足，则会被直接抛弃

df_of_calculating['tmp'] = 0

# list_fullfill = ['1_1']
# list_fullfill = ['1_1','1_2']
# list_fullfill = ['1_1', '1_2', '1_3', '1_4']
list_fullfill = df_of_product_sales.columns

for aim_month in list_fullfill:
    # 先算钱，成品1次到达仓库
    df_of_calculating['value_accumu'] = df_of_calculating['value_accumu'] + \
        df_of_method[aim_month] * 2 * df_of_method['unit feedstock value']  # 两周的零件仓储的value
    df_of_calculating['value_accumu'] = df_of_calculating['value_accumu'] + \
        df_of_calculating['stock'] * df_of_method['unit feedstock value']  # 目前所有的成品仓储的value

    # 目前存在到达成品和stock，优先使用stock，然后使用到达的成品。
    # 如果成品还有剩余，成品变为stock
    # 目前存在刚刚到达的需求，等待了1周的需求，等待了2周的需求；优先满足等待了2周的需求，然后满足等待1周的需求，最后满足刚刚到达的需求。
    # 如果需求未被满足，则数据延后一周
    
    # 计算卖出
    # 所有到达变成stock
    df_of_calculating['stock'] = df_of_calculating['stock'] + df_of_method[aim_month]
    
#     print(df_of_calculating)
    df_of_calculating['demand_today'] = df_of_product_sales[aim_month]# 临时变量，当前到达的需求
    
    
    # 目前存在需求 1.demand_today 2.demand_wait_1W 3.demand_wai_2W

    # 优先满足2W的需求
    df_of_calculating['tmp'] = df_of_calculating['stock']
    
    df_of_calculating['stock'] -= df_of_calculating['demand_wait_2W']
#     print(df_of_calculating["stock"])
    # 对于stock小于0的部分，将这些记录为未被满足的需求
    df_of_calculating.loc[df_of_calculating['stock']<0,'unfulfill_demand'] -= df_of_calculating.loc[df_of_calculating['stock'] < 0,'stock']
    
    # stock为负数的变为0，W2的未满足变为0，继续以下计算
    df_of_calculating.loc[df_of_calculating['stock'] < 0,'stock'] = 0
    df_of_calculating['fullfill_demand'] += (df_of_calculating['tmp'] - df_of_calculating['stock'])
    df_of_calculating['demand_wait_2W'] = 0
    
    
#     print(df_of_calculating["unfulfill_demand"])
#     print(df_of_calculating["fullfill_demand"])
#     print("stock 和 demand_wait_2W")
#     print(df_of_calculating['stock'])
#     print(df_of_calculating['demand_wait_2W'])
    
    
    # 其次满足1W的需求
    df_of_calculating['stock'] -= df_of_calculating['demand_wait_1W']
    # 对于stock小于0的部分，将这些记录挪到第2周
    df_of_calculating.loc[df_of_calculating['stock'] < 0,'demand_wait_2W'] -= df_of_calculating.loc[df_of_calculating['stock'] < 0,'stock']

    df_of_calculating['fullfill_demand'] += (df_of_calculating['demand_wait_1W'] - df_of_calculating['demand_wait_2W'])
    # stock为负数的变为0，W1的变为0，继续以下计算
    df_of_calculating.loc[df_of_calculating['stock'] < 0,'stock'] = 0
    df_of_calculating['demand_wait_1W'] = 0
    
#     print(df_of_calculating["unfulfill_demand"])
#     print(df_of_calculating["fullfill_demand"])    
#     print("stock 和 demand_wait_1W")
#     print(df_of_calculating['stock'])
#     print(df_of_calculating['demand_wait_1W'])
    

    
    # 满足today的需求
    df_of_calculating['stock'] -= df_of_calculating['demand_today']
    # 对于stock小于0的部分，将这些记录挪到第1周
    df_of_calculating.loc[df_of_calculating['stock']<0,'demand_wait_1W'] -= df_of_calculating.loc[df_of_calculating['stock'] < 0,'stock']

    df_of_calculating['fullfill_demand'] += (df_of_calculating['demand_today'] - df_of_calculating['demand_wait_1W'])
    # stock为负数的变为0
    df_of_calculating.loc[df_of_calculating['stock'] < 0,'stock'] = 0

    
#     print(df_of_calculating["unfulfill_demand"])
#     print(df_of_calculating["fullfill_demand"])
#     print("stock 和 demand_wait_1W")
#     print(df_of_calculating['stock'])
#     print(df_of_calculating['demand_wait_1W'])


#     print(df_of_calculating)
    print('↑'*15+aim_month+"finish"+'↑'*15)
df_of_calculating.drop(['tmp','demand_today','demand_wait_1W','demand_wait_2W'],axis = 1,inplace=True)
df_of_calculating['fulfill_rate'] = df_of_calculating['fullfill_demand']/(df_of_calculating['fullfill_demand']+df_of_calculating['unfulfill_demand'])
df_of_calculating.to_excel(OUTPUT_FILE_PATH3,sheet_name=SHEET_NAME)
df_of_calculating

# IELM6004
# Risk Scheduling
```
