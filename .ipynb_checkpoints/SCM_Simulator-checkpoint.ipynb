```python
# df_of_product_salesä»¿çœŸå‡ºçš„æ¯å‘¨éœ€æ±‚
# df_of_methodä»¿çœŸå‡ºçš„æ¯å‘¨ä¾›è´§

df_of_calculating = pd.DataFrame().reindex_like(df_of_product_sales).fillna(0)
df_of_calculating.drop(df_of_calculating.columns, axis=1, inplace=True)

df_of_calculating['value_accumu'] = 0  # è®¡ç®—æ‰€æœ‰ç‰©æ–™çš„ä»“å‚¨æˆæœ¬
df_of_calculating['stock'] = 0  # åº“å­˜æ•°é‡ï¼Œç”±äºŽorder-makeï¼Œæ‰€ä»¥æ²¡æœ‰è€ƒè™‘ä»¥é›¶ä»¶çš„æ–¹å¼è¿›è¡Œä»“å‚¨æƒ…å†µ

df_of_calculating['fullfill_demand'] = 0  # è®¡ç®—æ»¡è¶³äº†å¤šå°‘éœ€æ±‚ï¼Œæœªæ»¡è¶³çš„éœ€æ±‚å°±æŠ›å¼ƒ
df_of_calculating['unfullfill_demand'] = 0  # è®°å½•æœªè¢«æ»¡è¶³çš„éœ€æ±‚

df_of_calculating['demand_today'] = 0  # æ¯å‘¨ä¸€æ¥åˆ°çš„éœ€æ±‚
df_of_calculating['demand_wait_1W'] = 0  # ç­‰å¾…äº†1å‘¨çš„æœªè¢«æ»¡è¶³çš„éœ€æ±‚
df_of_calculating['demand_wait_2W'] = 0  # ç­‰å¾…äº†2å‘¨çš„æœªè¢«æ»¡è¶³çš„éœ€æ±‚ï¼Œå¦‚æžœç»§ç»­æœªè¢«æ»¡è¶³ï¼Œåˆ™ä¼šè¢«ç›´æŽ¥æŠ›å¼ƒ

df_of_calculating['tmp'] = 0

# list_fullfill = ['1_1']
# list_fullfill = ['1_1','1_2']
# list_fullfill = ['1_1', '1_2', '1_3', '1_4']
list_fullfill = df_of_product_sales.columns

for aim_month in list_fullfill:
    # å…ˆç®—é’±ï¼Œæˆå“1æ¬¡åˆ°è¾¾ä»“åº“
    df_of_calculating['value_accumu'] = df_of_calculating['value_accumu'] + \
        df_of_method[aim_month] * 2 * df_of_method['unit feedstock value']  # ä¸¤å‘¨çš„é›¶ä»¶ä»“å‚¨çš„value
    df_of_calculating['value_accumu'] = df_of_calculating['value_accumu'] + \
        df_of_calculating['stock'] * df_of_method['unit feedstock value']  # ç›®å‰æ‰€æœ‰çš„æˆå“ä»“å‚¨çš„value

    # ç›®å‰å­˜åœ¨åˆ°è¾¾æˆå“å’Œstockï¼Œä¼˜å…ˆä½¿ç”¨stockï¼Œç„¶åŽä½¿ç”¨åˆ°è¾¾çš„æˆå“ã€‚
    # å¦‚æžœæˆå“è¿˜æœ‰å‰©ä½™ï¼Œæˆå“å˜ä¸ºstock
    # ç›®å‰å­˜åœ¨åˆšåˆšåˆ°è¾¾çš„éœ€æ±‚ï¼Œç­‰å¾…äº†1å‘¨çš„éœ€æ±‚ï¼Œç­‰å¾…äº†2å‘¨çš„éœ€æ±‚ï¼›ä¼˜å…ˆæ»¡è¶³ç­‰å¾…äº†2å‘¨çš„éœ€æ±‚ï¼Œç„¶åŽæ»¡è¶³ç­‰å¾…1å‘¨çš„éœ€æ±‚ï¼Œæœ€åŽæ»¡è¶³åˆšåˆšåˆ°è¾¾çš„éœ€æ±‚ã€‚
    # å¦‚æžœéœ€æ±‚æœªè¢«æ»¡è¶³ï¼Œåˆ™æ•°æ®å»¶åŽä¸€å‘¨
    
    # è®¡ç®—å–å‡º
    # æ‰€æœ‰åˆ°è¾¾å˜æˆstock
    df_of_calculating['stock'] = df_of_calculating['stock'] + df_of_method[aim_month]
    
#     print(df_of_calculating)
    df_of_calculating['demand_today'] = df_of_product_sales[aim_month]# ä¸´æ—¶å˜é‡ï¼Œå½“å‰åˆ°è¾¾çš„éœ€æ±‚
    
    
    # ç›®å‰å­˜åœ¨éœ€æ±‚ 1.demand_today 2.demand_wait_1W 3.demand_wai_2W

    # ä¼˜å…ˆæ»¡è¶³2Wçš„éœ€æ±‚
    df_of_calculating['tmp'] = df_of_calculating['stock']
    
    df_of_calculating['stock'] -= df_of_calculating['demand_wait_2W']
#     print(df_of_calculating["stock"])
    # å¯¹äºŽstockå°äºŽ0çš„éƒ¨åˆ†ï¼Œå°†è¿™äº›è®°å½•ä¸ºæœªè¢«æ»¡è¶³çš„éœ€æ±‚
    df_of_calculating.loc[df_of_calculating['stock']<0,'unfullfill_demand'] -= df_of_calculating.loc[df_of_calculating['stock'] < 0,'stock']
    
    # stockä¸ºè´Ÿæ•°çš„å˜ä¸º0ï¼ŒW2çš„æœªæ»¡è¶³å˜ä¸º0ï¼Œç»§ç»­ä»¥ä¸‹è®¡ç®—
    df_of_calculating.loc[df_of_calculating['stock'] < 0,'stock'] = 0
    df_of_calculating['fullfill_demand'] += (df_of_calculating['tmp'] - df_of_calculating['stock'])
    df_of_calculating['demand_wait_2W'] = 0
    
    
#     print(df_of_calculating["unfullfill_demand"])
#     print(df_of_calculating["fullfill_demand"])
#     print("stock å’Œ demand_wait_2W")
#     print(df_of_calculating['stock'])
#     print(df_of_calculating['demand_wait_2W'])
    
    
    # å…¶æ¬¡æ»¡è¶³1Wçš„éœ€æ±‚
    df_of_calculating['stock'] -= df_of_calculating['demand_wait_1W']
    # å¯¹äºŽstockå°äºŽ0çš„éƒ¨åˆ†ï¼Œå°†è¿™äº›è®°å½•æŒªåˆ°ç¬¬2å‘¨
    df_of_calculating.loc[df_of_calculating['stock'] < 0,'demand_wait_2W'] -= df_of_calculating.loc[df_of_calculating['stock'] < 0,'stock']
    # è®°å½•æ»¡è¶³çš„éœ€æ±‚(å…¶å®žå¯ä»¥é€šè¿‡æœ€åŽæœªæ»¡è¶³å–åï¼Œä½†æ˜¯èµ·ä¸€ä¸ªéªŒè¯çš„ä½œç”¨è¿˜æ˜¯é€šè¿‡è¿™ç§æ–¹æ³•ç»§ç»­è®¡ç®—)
    df_of_calculating['fullfill_demand'] += (df_of_calculating['demand_wait_1W'] - df_of_calculating['demand_wait_2W'])
    # stockä¸ºè´Ÿæ•°çš„å˜ä¸º0ï¼ŒW1çš„å˜ä¸º0ï¼Œç»§ç»­ä»¥ä¸‹è®¡ç®—
    df_of_calculating.loc[df_of_calculating['stock'] < 0,'stock'] = 0
    df_of_calculating['demand_wait_1W'] = 0
    
#     print(df_of_calculating["unfullfill_demand"])
#     print(df_of_calculating["fullfill_demand"])    
#     print("stock å’Œ demand_wait_1W")
#     print(df_of_calculating['stock'])
#     print(df_of_calculating['demand_wait_1W'])
    

    
    # æ»¡è¶³todayçš„éœ€æ±‚
    df_of_calculating['stock'] -= df_of_calculating['demand_today']
    # å¯¹äºŽstockå°äºŽ0çš„éƒ¨åˆ†ï¼Œå°†è¿™äº›è®°å½•æŒªåˆ°ç¬¬1å‘¨
    df_of_calculating.loc[df_of_calculating['stock']<0,'demand_wait_1W'] -= df_of_calculating.loc[df_of_calculating['stock'] < 0,'stock']
    # è®°å½•æ»¡è¶³çš„éœ€æ±‚
    df_of_calculating['fullfill_demand'] += (df_of_calculating['demand_today'] - df_of_calculating['demand_wait_1W'])
    # stockä¸ºè´Ÿæ•°çš„å˜ä¸º0
    df_of_calculating.loc[df_of_calculating['stock'] < 0,'stock'] = 0

    
#     print(df_of_calculating["unfullfill_demand"])
#     print(df_of_calculating["fullfill_demand"])
#     print("stock å’Œ demand_wait_1W")
#     print(df_of_calculating['stock'])
#     print(df_of_calculating['demand_wait_1W'])


#     print(df_of_calculating)
    print('â†‘'*30+aim_month+"finish"+'â†‘'*30)

df_of_calculating
# IELM6004
# Risk Scheduling
```
